```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd


url = "https://home.instacard.co/examples/"


driver = webdriver.Chrome()  


driver.get(url)


anchor_tags = driver.find_elements(By.CSS_SELECTOR, 'a.ct-link')


unique_urls = set()


data_list = []


for i, anchor in enumerate(anchor_tags):
    if i < 6 or i >= len(anchor_tags) - 2:
        continue

    url = anchor.get_attribute('href')
    if url and url not in unique_urls:
        unique_urls.add(url)
        data_list.append({"URL": url})


df = pd.DataFrame(data_list)


df.to_csv("store_link.csv", index=False)


def scrape_content(url):
    try:
       
        driver.get(url)

       
        driver.implicitly_wait(10)

       
        title = driver.find_element(By.TAG_NAME, "h1").text if driver.find_element(By.TAG_NAME, "h1").text else "N/A"
        phone = driver.find_element(By.CLASS_NAME, "agent-phone").text if driver.find_element(By.CLASS_NAME, "agent-phone").text else "N/A"
        description = driver.find_element(By.TAG_NAME, "p").text if driver.find_element(By.TAG_NAME, "p").text else "N/A"
        images = [img.get_attribute("src") for img in driver.find_elements(By.TAG_NAME, "img")]
        email = driver.find_element(By.CLASS_NAME, "ant-email").text if driver.find_elements(By.CLASS_NAME, "ant-email") else "N/A"
        company = driver.find_element(By.CLASS_NAME, "ant-company").text if driver.find_elements(By.CLASS_NAME, "ant-company") else "N/A"

       
        print("URL:", url)
        print("Title:", title)
        print("Phone:", phone)
        print("Email:", email)
        print("Description:", description)
        print("Images:", images)
        print("Company:", company)
        print("\n")

    except Exception as e:
        print(f"Error scraping {url}: {str(e)}")


df = pd.read_csv("store_link.csv")
for url in df["URL"]:
    scrape_content(url)


driver.quit()
```
